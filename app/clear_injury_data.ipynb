# Prerequizets

### Remove non-sport related injuries
NEISS tables represent all types of injuries in the US. So before we start with with our resurch it's important to remove all of the injuries that are not related to sport at all

```python
table_2024 = pd.read_csv('../data/original_data/injury_data/NEISS_2024.csv')
table_2023 = pd.read_csv('../data/original_data/injury_data/NEISS_2023.csv')
table_2022 = pd.read_csv('../data/original_data/injury_data/NEISS_2022.csv')
table_2021 = pd.read_csv('../data/original_data/injury_data/NEISS_2021.csv')
table_2020 = pd.read_csv('../data/original_data/injury_data/NEISS_2020.csv')
table_2019 = pd.read_csv('../data/original_data/injury_data/NEISS_2019.csv')
table_2018 = pd.read_csv('../data/original_data/injury_data/NEISS_2018.csv')
table_2017 = pd.read_csv('../data/original_data/injury_data/NEISS_2017.csv')
table_2016 = pd.read_csv('../data/original_data/injury_data/NEISS_2016.csv')
table_2015 = pd.read_csv('../data/original_data/injury_data/NEISS_2015.csv')

injury_tables = [
    table_2024, table_2023, 
    table_2022, table_2021, 
    table_2020, table_2019, 
    table_2018, table_2017, 
    table_2016, table_2015
    ]

mapping_table = pd.read_csv('../data/extracted_data/codes/edited_sport_codes.csv')
```

```python
filtered_tables = bulk_filter_by_code(
    main_tables_list=injury_tables, mapping_table=mapping_table, 
    target_cols_list=['Product_1', 'Product_2', 'Product_3']
    )
```

### Replace codes with meaning
Because the NEISS tables columns contain mainly numbers that represen code, we will replace them with text so it will be more convinient for us to work like this

1. Product codes

```python
product_codes_table = mapping_table

meaning_replaced_tables = bulk_replace_codes_with_meanings(
    main_tables_list=filtered_tables, mapping_table=product_codes_table,
    target_cols=['Product_1', 'Product_2', 'Product_3'], meaning_col='Sport'
    )
```

2. Gender

```python
gender_codes_table = pd.read_csv('../data/original_data/codes/NEISS_tables/gender_codes.csv')

meaning_replaced_tables = bulk_replace_codes_with_meanings(
    main_tables_list=meaning_replaced_tables, mapping_table=gender_codes_table,
    target_cols=['Sex'], meaning_col='Gender'
    )
```

3. Race

```python
race_codes_table = pd.read_csv('../data/original_data/codes/NEISS_tables/race_codes.csv')

meaning_replaced_tables = bulk_replace_codes_with_meanings(
    main_tables_list=meaning_replaced_tables, mapping_table=race_codes_table,
    target_cols=['Race'], meaning_col='Race'
    )
```

4. Hispanic

```python
yes_no_codes_table = pd.read_csv('../data/original_data/codes/NEISS_tables/yes_no_codes.csv')

meaning_replaced_tables = bulk_replace_codes_with_meanings(
    main_tables_list=meaning_replaced_tables, mapping_table=yes_no_codes_table,
    target_cols=['Hispanic'], meaning_col='Is statement true'
    )
```

5. Body part

```python
yes_no_codes_table = pd.read_csv('../data/original_data/codes/NEISS_tables/yes_no_codes.csv')

meaning_replaced_tables = bulk_replace_codes_with_meanings(
    main_tables_list=meaning_replaced_tables, mapping_table=yes_no_codes_table,
    target_cols=['Hispanic'], meaning_col='Is statement true'
    )
```
