12. Age

Important clarifications
- Every person above 115 years of age is represented as 115 years old.
- Eveyone under 2 years of age is represent in range of 200-223
    - The last 2 digits represent the age in months

```python
def clean_age_numeric(val):
    """
    Converts months to fraction of a year. 
    Ensure that diagrams will be represented correctly later on.
    """
    if 200 <= val <= 223:
        return (val - 200) / 12 
    return float(val)
```

```python
for table in meaning_replaced_tables:
    table["Age"] = table["Age"].apply(clean_age_numeric)
```
